**Radioactive Decay Simulation**  

This notebook simulates the exponential decay of a radioactive nucleus using numerical methods and visualizes the decay process.


 **1. Introduction**  

Radioactive decay follows an exponential law:  
\[
N(t) = N_0 e^{-\lambda t}
\]  
where:  
- \( N(t) \) = Number of nuclei at time \( t \)  
- \( N_0 \) = Initial number of nuclei  
- \( \lambda \) = Decay constant (\( \lambda = \frac{\ln(2)}{t_{1/2}} \))  
- \( t_{1/2} \) = Half-life  

 **2. Numerical Solution**  

We simulate decay using both exact solutions and stochastic methods.


# **Import Libraries**  

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
import random
````

# **Exact Solution (Analytical)**  

````python

def exact_decay(N0, decay_const, t):
    return N0 * np.exp(-decay_const * t)
````

# **Stochastic Simulation (Monte Carlo)**  

````python

def stochastic_decay(N0, decay_const, t_max, dt):
    times = np.arange(0, t_max, dt)
    N = np.zeros_like(times)
    N[0] = N0

    for i in range(1, len(times)):
        decay_prob = 1 - np.exp(-decay_const * dt)
        N[i] = N[i-1] - np.sum(np.random.rand(N[i-1]) < decay_prob)

    return times, N
````
